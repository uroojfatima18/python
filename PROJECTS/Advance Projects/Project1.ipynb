# **HIGH LOW GAME**

```python
num_round = 5
round_num = 0
score = 0
import random
print('Welcome to High Low Game....')

# Milestone 4: Play multiple rounds
while round_num < num_round:
    print("--------------------------------")
    round_num += 1
    print(f"Round {round_num}")
    print("--------------------------------")

#  Milestone 1 :Generate the random numbers
    comp_num = random.randint(1, 100)
    user_num = random.randint(1, 100)

    print(f"Your Number is: {user_num}")
    # print(f"The Computer Number is: {comp_num}")

#  Milestone 2 :Get the user choice
    guess = str(input("Do you think your number is high or low than the computer's? (high/low):")).lower().strip()
    while guess not in ["high","low"]:
      print("Invalid input!Please enter either 'high' or 'low'.")
      guess = str(input("Do you think your number is high or low than the computer's? (high/low):")).lower().strip()

#  Milestone 3 :Write the game logic
    if guess == "high":
        if user_num > comp_num:
            print(f"You were right! The computer number is: {comp_num}")
# Milestone 5: Adding a points system
            score += 1  # Increment score if the guess is correct
        else:
            print(f"Aww, that's incorrect. The computer number is: {comp_num}")
    elif guess == "low":
        if user_num < comp_num:
            print(f"You were right! The computer number is: {comp_num}")
            score += 1  # Increment score if the guess is correct
        else:
            print(f"Aww, that's incorrect. The computer number is: {comp_num}")

    print(f"Your current score is: {score}/{num_round}")

print(f"Your Final Score: {score}/{num_round}")
if score == num_round:
    print("Wow! You played perfectly!")
elif score >= num_round // 2:
    print("Good job, you played really well!")
else:
    print("Better luck next time!")
```
